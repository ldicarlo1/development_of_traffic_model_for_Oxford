# Create Road-Network Graph

```python
# load in csv of node connections
connections = pd.read_csv("peartree_roundabout.csv")
connections.head(5)
```

```python
# convert feeding roads to integers
for i in range(len(connections)):
#for i in range(4):
    try:
        connections.feeding_roads.iloc[i] = ast.literal_eval(connections.feeding_roads.iloc[i])
    except ValueError:
        connections.feeding_roads.iloc[i] = np.nan
```

Collect nodes for Graph 

```python
# node connections
#nodes = connections["Unnamed: 0"][:4]
#roads = connections.feeding_roads[:4]
nodes = connections["Unnamed: 0"]
roads = connections.feeding_roads
# replace nans with 0's
connections.feeding_roads = connections.feeding_roads.fillna(0)
```

```python
# loop thru and establish edges
edge_list = []
for row in range(len(roads)):
    node1 = connections["Unnamed: 0"].iloc[row]
    node2 = connections.feeding_roads.iloc[row]
    try:
        for i in range(len(node2)):
            edge_list.append([node2[i], node1])
        #node2 = connections.feeding_roads.iloc[row]
    except TypeError:
        edge_list.append([node2, node1])
        
# remove 0's
edges = []
for edge in edge_list:
    if edge[0]==0:
        pass
    else:
        edges.append(edge)
        
```

```python
#build the graph
G = nx.Graph()
for i in range(len(nodes)):
    G.add_node(nodes[i],spd=sp[:,i])
    #G.add_edge(nodes[i])
#G.add_nodes_from(nodes)
G.add_edges_from(edges)

```

```python
plt.figure(1,figsize=(8,6)) 
nx.draw(G,with_labels = False,node_size=200,pos=nx.spring_layout(G),node_color="skyblue")
```

# Graph Convolutional Neural Network

```python
# convert graph to stellargraph object for modeling
square = StellarGraph.from_networkx(G,node_features="spd")

# get feature matrix
X = square.node_features()
```

```python
X = np.concatenate([X,X,X,X,X,X,X],axis=1)
```

```python
#generator = sg.mapper.FullBatchNodeGenerator(square, method="gcn")
#generator
```

```python
def train_test_split(data, train_portion):
    time_len = data.shape[1]
    train_size = int(time_len * train_portion)
    train_data = np.array(data[:, :train_size])
    test_data = np.array(data[:, train_size:])
    return train_data, test_data
```

```python
train_rate = 0.8
```

```python
X = np.where(X<0,0,X)
train_data, test_data = train_test_split(X, train_rate)
print("Train data: ", train_data.shape)
print("Test data: ", test_data.shape)
```

```python
def scale_data(train_data, test_data):
    max_speed = train_data.max()
    min_speed = train_data.min()
    train_scaled = (train_data - min_speed) / (max_speed - min_speed)
    test_scaled = (test_data - min_speed) / (max_speed - min_speed)
    return train_scaled, test_scaled

train_scaled, test_scaled = scale_data(train_data, test_data)
```

```python
seq_len = 10
pre_len = 12
```

```python
def sequence_data_preparation(seq_len, pre_len, train_data, test_data):
    trainX, trainY, testX, testY = [], [], [], []

    for i in range(train_data.shape[1] - int(seq_len + pre_len - 1)):
        a = train_data[:, i : i + seq_len + pre_len]
        trainX.append(a[:, :seq_len])
        trainY.append(a[:, -1])

    for i in range(test_data.shape[1] - int(seq_len + pre_len - 1)):
        b = test_data[:, i : i + seq_len + pre_len]
        testX.append(b[:, :seq_len])
        testY.append(b[:, -1])

    trainX = np.array(trainX)
    trainY = np.array(trainY)
    testX = np.array(testX)
    testY = np.array(testY)

    return trainX, trainY, testX, testY
```

```python
trainX, trainY, testX, testY = sequence_data_preparation(
    seq_len, pre_len, train_scaled, test_scaled
)
print(trainX.shape)
print(trainY.shape)
print(testX.shape)
print(testY.shape)
```

```python
gcn_lstm = GCN_LSTM(
    seq_len=seq_len,
    adj=A,
    gc_layer_sizes=[16,10],
    gc_activations=["relu","relu"],
    lstm_layer_sizes=[200,200],
    lstm_activations=["tanh","tanh"],
)
```

```python
x_input, x_output = gcn_lstm.in_out_tensors()
```

```python
model = Model(inputs=x_input, outputs=x_output)
```

```python
model.compile(optimizer="adam", loss="mae", metrics=["mse"])
model.summary()
```

```python
history = model.fit(
    trainX,
    trainY,
    epochs=70,
    shuffle=True,
    batch_size=60,
    verbose=1,
    validation_data=(trainX,trainY)
)
```

```python
sg.utils.plot_history(history)
```
