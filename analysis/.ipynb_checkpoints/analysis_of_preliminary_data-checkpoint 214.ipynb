# Preliminary Data Analysis

###### This data analysis is using the preliminary data to gauge only June 23, 2021 on 5-minute intervals. This is to provide a basic understanding of the daily cycle of traffic. 

```python
import pandas as pd 
import numpy as np 
from datetime import datetime, timedelta
import plotly.express as px
import ast
import matplotlib.pyplot as plt 
```

```python
def subset_data(data):
    # Peartree Roundabout bounding-box
    top=51.798433
    bottom=51.791451
    right=-1.281979
    left=-1.289524
    
    loni = []
    lati=[]
    for i in range(len(traffic_matrix)):
        lati.append(ast.literal_eval(data.lats.iloc[i]))
        loni.append(ast.literal_eval(data.lons.iloc[i]))

    data['loni'] = loni
    data['lati'] = lati
    bbox_data=[]
    for i in range(len(data)):
        lonmin = np.array(data.loni.iloc[i]).min()
        lonmax = np.array(data.loni.iloc[i]).max()
        latmin = np.array(data.lati.iloc[i]).min()
        latmax = np.array(data.lati.iloc[i]).max()
        if latmax<=top and latmin >=bottom and lonmin >= left and lonmax <=right:
            bbox_data.append(data.iloc[i].values)
    bbox_data = pd.DataFrame(np.array(bbox_data))
    bbox_data.columns = data.columns
    
    return bbox_data
```

```python
def load_prelim_traffic_data(datetime_start, datetime_end):
    
    '''
    This function collects the preliminary road network traffic data for Oxford, UK on 5-minute intervals from 
    the specified start and end datetimes.
    
    Args: 
    
    datetime_start (datetime object): specifies the start datetime for creating a range of datetimes
    
    datetime_end (datetime object): specifies the end datetime for creating a range of datetimes
    
    Returns:
    
    3d-array (array_obj): dims (time, number of road segments, columns)

     
    '''
    
    # empty traffic/time dataframe to be filled
    traffic_data = []

    # create date-time range for all file times
    time = pd.date_range(datetime_start,datetime_end,freq="5min")
    datetimestamps = time.strftime("%Y%m%d%H%M")
    
    # create array with shape (time, number of road segments, columns)
    for datetimestamp in datetimestamps:
        #print(datetimestamp)
        dir = "/Users/lucadicarlo/Documents/OxfordBrookes/dissertation/development_of_traffic_model_for_Oxford/data_collection/data/traffic_data_20210623_prelim/"
        filename = f"oxford_traffic_{datetimestamp}.csv"
        traffic_matrix = pd.read_csv(dir+filename)
        
        
        # subset data for Peartree Roundabout
        traffic_matrix_subset = subset_data(traffic_matrix)
        
        traffic_data.append(traffic_matrix_subset)
    traffic_data = np.array(traffic_data)
    
    
    # Peartree Roundabout bounding-box
    top=51.798433
    bottom=51.791451
    right=-1.281979
    left=-1.289524
    
    
    
    

    return traffic_data
```

```python
# load traffic data for prelim
datetime_start=datetime(2021,6,23,0,0)
datetime_end=datetime(2021,6,23,23,55)

# call load data function
traffic_data = load_prelim_traffic_data(datetime_start,datetime_end)
```

### Mean Traffic for the Peartree Roundabout

```python
# pull traffic metrics average for city
sp = traffic_data[:,:,5].mean(axis=1)
su = traffic_data[:,:,6].mean(axis=1)
ffs = traffic_data[:,:,7].mean(axis=1)
jf = traffic_data[:,:,8].mean(axis=1)
lons = traffic_data[0,:,4]
lats = traffic_data[0,:,3]

flow = sp/ffs
```

### Geographic Plots

```python
# select time to plot traffic
ind = 90
bbox_data = traffic_data

# calculate flow by dividing speed by normal speed
flow = bbox_data[:,:,5].mean(axis=0)/bbox_data[:,:,7].mean(axis=0)

# collect time
geo_time = time[ind].strftime("%d %B, %Y")
print(geo_time)
```

```python
from matplotlib.backends.backend_agg import FigureCanvas
colors = plt.cm.gray(np.linspace(0,1,256))
colors= colors[:,:3]
colors = np.round(colors,8)


# Create geographic plot of road network 
fig=plt.figure(dpi=50)
#plt.plot(np.linspace(0,10,10),np.linspace(0,10,10))
plt.rcParams["figure.figsize"] = (10,10)
plt.style.use('dark_background')
plt.grid(False)
for i in range(0,len(flow)):
    plt.xlim(-1.281979,-1.289524)
    if(flow[i]<0.25):
        plt.plot(ast.literal_eval(lons[i]),ast.literal_eval(lats[i]), c="red",linewidth=4)
    elif(flow[i]<0.5):
        plt.plot(ast.literal_eval(lons[i]),ast.literal_eval(lats[i]), color="orange",linewidth=4)
    elif(flow[i]<0.75):
        plt.plot(ast.literal_eval(lons[i]),ast.literal_eval(lats[i]), c='yellow',linewidth=4)
    else:
        plt.plot(ast.literal_eval(lons[i]),ast.literal_eval(lats[i]), c='green',linewidth=4)
    

plt.axis('off')
plt.title(f"Daily Average Peartree Roundabout Traffic Flow: {geo_time}")
plt.show()

        
```
